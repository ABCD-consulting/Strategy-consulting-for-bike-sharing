```python
import pandas as pd
from geopy.geocoders import Nominatim

station_data = pd.read_csv('austin-bike/austin_bikeshare_stations.csv')
station_data[:10]
```

```python
import certifi
import ssl
import geopy.geocoders
ctx = ssl.create_default_context(cafile=certifi.where())
geopy.geocoders.options.default_ssl_context = ctx
```

```python
", ".join(station_data["location"][0][1:-1].split(" "))
```

```python
geolocator = Nominatim()
geolocator
```

```python
station_data["zipcode"] = station_data["location"].map( 
   lambda x : str(geolocator.reverse(x[1:-1].replace(" ", ", ")).address.split(", ")[-2]) )
```

```python
station_data.to_csv("austin_bikeshare_stations_w_zipcode.csv")
```
